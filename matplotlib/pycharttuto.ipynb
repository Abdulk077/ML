```python
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
exp_value = [1600,200,500,600,120]
exp_lables = ['Home Rent ', 'Food', 'phone' , 'car', 'other utility']
```

```python
plt.axis("equal")
plt.title("pie chart")
plt.pie(exp_value,labels=exp_lables ,autopct='%0.1f%%' , radius=1.5 , shadow=True, explode=[0,0,0,0.2,0])
plt.savefig("piechart.png",bbox_inches="tight", pad_inches=2, transparent=True)
```
